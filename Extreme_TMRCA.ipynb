Il semblerait que $\delta_0$ ne soit pas la mesure qui maximise $\mathbb E(TMRCA)$.
Regardons ça de plus près.

Prenons $n=8$, alors,

$$\mathbb E(TMRCA) = \sum_{k=2}^{8} \frac{2}{k(k-1)} = 2\left(1 - \frac{1}{8}\right) = \frac{14}{8} = 1.75$$

A présent trouvons une mesure qui donne une valeur plus grande.
Soit $\Lambda = \delta_{1/4}$.
Alors,

$$
\lambda_{n,k} = \int_0^1 x^{k-2}(1-x)^{n-k} d\Lambda(x) = \left(\frac{1}{4}\right)^{k-2} \left(1 - \frac{1}{4}\right)^{n-k} = \left(\frac{1}{4}\right)^{k-2} \left(\frac{3}{4}\right)^{n-k}
$$


$$
\lambda_n = \sum_{k=2}^{n} \binom{n}{k} \lambda_{n,k} = \sum_{k=2}^{n} \binom{n}{k} \left(\frac{1}{4}\right)^{k-2} \left(\frac{3}{4}\right)^{n-k} =  16\left(1 -\left(\frac34\right)^n - \frac{n}{4}\left(\frac34\right)^{n-1}\right)
$$


$$\mathbb E(TMRCA) = \mathbb E(\tau |N_0 = n)  = \frac{1}{\lambda_n} + \sum_{k=2}^{n} \frac{\binom{n}{k} \lambda_{n,k}}{\lambda_n} \mathbb E(\tau | N_0 = n-k+1 )$$

On a $\mathbb E(\tau | N_0 = 1) = 0, \mathbb E(\tau | N_0 = 2) = 1$.

```python
import numpy as np
import math
```

```python
n=8
lmb_n_k = lambda n, k : 0.25**(k-2) * (0.75)**(n - k) 
lmb_n = lambda n : 16 * (1 - (0.75)**n - n/4 * (0.75)**(n-1))
print(lmb_n(n))
```

```python
Esperance = [0, 1]
for i in range(3, n+1):
    En = 1/lmb_n(i) + np.sum([
        math.comb(i,k)*(lmb_n_k(i,k))/lmb_n(i) * Esperance[i-k+1 -1]  for k in range(2,i)
    ])
    Esperance.append(En)
```
